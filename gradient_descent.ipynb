```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the cost function and its derivative
def cost_function(x):
    return x**2

def gradient(x):
    return 2 * x
```

```python
# Gradient Descent function
def gradient_descent(learning_rate, num_iterations):
    x = 10  # Initial guess for the minimum
    x_history = [x]

    for i in range(num_iterations):
        grad = gradient(x)
        x -= learning_rate * grad
        x_history.append(x)

    return x_history
```

```python
# Parameters for gradient descent
learning_rate = 0.1
num_iterations = 50
```

```python
# Run gradient descent
history = gradient_descent(learning_rate, num_iterations)
```

```python
# Plot the convergence
plt.plot(history, cost_function(np.array(history)), 'bo-')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Convergence')
plt.show()
```
